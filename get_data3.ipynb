```python
headers = {
'Accept': '*/*',
'Accept-Encoding': 'gzip, deflate, br, zstd',
'Accept-Language': 'en-US,en;q=0.9,he;q=0.8',
'Connection': 'keep-alive',
'Host': 'stats.nba.com',
'Origin': 'https://www.nba.com',
'Referer': 'https://www.nba.com/',
'Sec-Ch-Ua': '"Chromium";v="122", "Not(A:Brand";v="24", "Google Chrome";v="122"',
'Sec-Ch-Ua-Mobile': '?1',
'Sec-Ch-Ua-Platform': '"Android"',
'Sec-Fetch-Dest': 'empty',
'Sec-Fetch-Mode': 'cors',
'Sec-Fetch-Site': 'same-site',
'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Mobile Safari/537.36'
}

df_cols = ['Year', 'Season_type']
df= pd.DataFrame(columns=df_cols)
season_types = ['Regular Season', 'Playoffs']
years = [f'{i}-{i-2000+1}' for i in range(2012, 2022)]

begin_loop = time.time()
for y in years:
    for s in season_types:
        season_type_encoded = s.replace(' ', '%20')
        api_url = f'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=S&Season={y}&SeasonType={season_type_encoded}&StatCategory=PTS'
        response = requests.get(url=api_url, headers=headers)
        
        if response.status_code == 200:
            r = response.json()
            temp_df1 = pd.DataFrame(r['resultSet']['rowSet'], columns=table_headers)
            data = {'Year': [y for _ in range(len(temp_df1))], 'Season_type': [s for _ in range(len(temp_df1))]}
            temp_df2 = pd.DataFrame(data)
            temp_df3 = pd.concat([temp_df2, temp_df1], axis=1)
            df = pd.concat([df, temp_df3], axis=0)
            print(f'Finished scraping for {y} {s}')
            
            # Sleep for a random time
            lag = np.random.uniform(5, 10)  # Adjusted for a shorter wait time
            time.sleep(lag)
            
            elapsed_time = round((time.time() - begin_loop) / 60, 2)
            print(f'Completed {y} {s} in {elapsed_time} mins')
        else:
            print(f'Failed to fetch data for {y} {s} with status code {response.status_code}')

# Save the compiled DataFrame to an Excel file
df.to_excel('nba_stats.xlsx', index=False)
```
